```python
def next_monday(date):
    return date + pd.Timedelta(days=(7 - date.weekday()))

def add_processing_time(order_date):
    if order_date.weekday() >= 5:  # weekend
        start = next_monday(order_date)
    else:
        start = order_date
    return start + pd.Timedelta(days=2)
```

```python
def next_truck_day(date):
    weekday = date.weekday()
    
    if weekday in [0, 2, 4]:  # Mon, Wed, Fri
        return date
    elif weekday == 1:  # Tue → Wed
        return date + pd.Timedelta(days=1)
    elif weekday == 3:  # Thu → Fri
        return date + pd.Timedelta(days=1)
    elif weekday == 5:  # Sat → Mon
        return date + pd.Timedelta(days=2)
    elif weekday == 6:  # Sun → Mon
        return date + pd.Timedelta(days=1)
```

```python
def calculate_delivery(row):
    
    # Step 1: Ready to ship
    ready_date = add_processing_time(row['Order Date'])
    
    # Step 2: Express (First_class == 1)
    if row['First_class'] == 1:
        ship_date = ready_date
    
    else:
        # wait 1 day
        ship_date = ready_date + pd.Timedelta(days=1)
        
        # move to next truck day
        ship_date = next_truck_day(ship_date)
    
    # Step 3: Delivery (same day)
    return ship_date
```

```python
Merge_date['delivery_date'] = Merge_date.apply(calculate_delivery, axis=1)
```

```python
Merge_date['shipping_time'] = (Merge_date['delivery_date'] - Merge_date['Pickup Date']).dt.days
```

```python
Merge_date['total_delivery_time'] = (Merge_date['delivery_date'] - Merge_date['Order Date']).dt.days
```
